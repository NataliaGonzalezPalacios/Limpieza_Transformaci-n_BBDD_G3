```python
## Librerías

import pandas as pd
import numpy as np
```

```python
from pandas.io.excel import read_excel
# Carpeta de Bases de datos https://github.com/Neo3009/BasesdeDatosgrupo2

url ="https://raw.githubusercontent.com/Neo3009/BasesdeDatosgrupo2/main/final.csv"
df0 = pd.read_csv(url)
df0
```

## Back Up del archivo paises en github
"https://github.com/LarryTobonS/Files/blob/main/Paises.xlsx"

```python
## Importar tabla de paises para generar códigos
tps = read_excel("Paises.xlsx",sheet_name="Codigos")
tps
```

```python
## Traer el código del país al df0

df1 = pd.merge(df0,tps,how="left",left_on="country",right_on="country")

## Ordenar columnas para que Code salga de primero
df1 = df1[['Code','country', 'region', 'health_index', 'obese_rate', 'suicide_rate',
       'democracies_rate', 'hdr_rate', 'life_Exp', 'gpi', 'iq_rate',
       'Deaths per 1,000 live births', 'gdp_usd', 'Ipc']]

df1
```

```python
#Filtrar los registros de Latinoamerica y generar base final
df = df1.loc[df1["region"] == "LATIN AMER. & CARIB"]
df
```

```python
## Exportar archivo transformado a CSV
df.to_csv('Datos_Macro.csv', index=False)
```
