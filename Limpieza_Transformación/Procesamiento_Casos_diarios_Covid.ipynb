##Actualización diarios de casos archivados

```python
import os
import glob
import pandas as pd
os.chdir("/content/drive/MyDrive/Grupo3DB/archived_daily_case_updates")
```

```python
extension = 'csv'
todos_los_archivos = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
total_casos = pd.concat([pd.read_csv(f) for f in todos_los_archivos ])
```
