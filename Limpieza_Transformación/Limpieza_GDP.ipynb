### Nos quedamos con los últimos 10 años y solo con los paises latinoamericanos

```python
bbdd_2=bbdd.set_index('GDP, current prices (Billions of U.S. dollars)')
bbdd_2 = bbdd_2.iloc[:,-11:]
bbdd_2 = bbdd_2.reset_index()
bbdd_2.head()
```

```python
paises = pd.read_excel('Paises_Lat.xlsx')
paises.head()
```

```python
bbdd_2['GDP, current prices (Billions of U.S. dollars)'].unique()
```

### Cruzando las bases

```python
bbdd_3 = bbdd_2.merge(paises ,how='left', left_on= ['GDP, current prices (Billions of U.S. dollars)'], right_on = 'País' )
print(bbdd_2.shape)
print(bbdd_3.shape)
```

```python
bbdd_3[bbdd_3['País'].notna()].shape
```

```python
bbdd_3[bbdd_3['País'].notna()]
```

### Me quedo solo con los paises latinoamericanos

```python
bbdd_4 = bbdd_3[bbdd_3['País'].notna()][['Código',  '2013', '2014',
       '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023']].copy()
```

### Validando tipo de dato

```python
bbdd_4.dtypes
```

```python
bbdd_4['Código'] = bbdd_4['Código'].astype('int64')
```

```python
bbdd_4.dtypes
```

### Renombrando las variables 

```python
bbdd_4.rename(columns={'Código':'codigo_pais'}, inplace= True)
```

### Validado duplicados

```python
#Check por duplicados 
duplicados= bbdd_4[bbdd_4.duplicated(['codigo_pais'], keep=False)] 
display('La cantidad de créditos duplicados es: ' + str(duplicados.shape[0]))
```

### Descargando base final

```python
bbdd_4.to_excel('bbdd_GDP.xlsx', index= False)
```
