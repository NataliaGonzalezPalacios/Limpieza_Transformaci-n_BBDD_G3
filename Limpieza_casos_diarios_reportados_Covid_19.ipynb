### Cargando Librerias

```python
import pandas as pd
import numpy as np
import glob
import os
import datetime as dt
from datetime import datetime
```

### Cargando la base de datos

```python
os.chdir(r'C:\Users\NATALIA\Documents\MAESTRIA\Base_datos')
paises = pd.read_excel('Paises_Lat.xlsx')
lista_paises = paises.País.unique().tolist()
paises.head()
```

#### Nos quedamos únicamnete con los paises latinoamericanos

```python
os.chdir(r'C:\Users\NATALIA\Documents\MAESTRIA\Base_datos\csse_covid_19_daily_reports')
filenames = glob.glob('***.csv*')
#print(filenames)
print(len(filenames))
base_casos_diarios = pd.DataFrame([])
for file in filenames:
    print(file)
    pfile = pd.read_csv(file)
    pfile.columns=pfile.columns.str.replace('/', '_')
    pfile = pfile[pfile['Country_Region'].isin(lista_paises)].copy()
    pfile['fecha'] = file[0:10]
    print(pfile.shape)
    base_casos_diarios = pd.concat([base_casos_diarios, pfile], sort= True)
base_casos_diarios.shape
```

#### Unificando columnas

```python
base_casos_diarios['Latitude']= np.where(base_casos_diarios['Latitude'].isna(), base_casos_diarios['Lat'], base_casos_diarios['Latitude'])
base_casos_diarios['Longitude']= np.where(base_casos_diarios['Longitude'].isna(), base_casos_diarios['Long_'], base_casos_diarios['Longitude'])
base_casos_diarios['Last_Update'] = np.where(base_casos_diarios['Last_Update'].isna(), base_casos_diarios['Last Update'], base_casos_diarios['Last_Update'])
base_casos_diarios['Case_Fatality_Ratio'] = np.where(base_casos_diarios['Case_Fatality_Ratio'].isna(), base_casos_diarios['Case-Fatality_Ratio'], base_casos_diarios['Case_Fatality_Ratio'])
base_casos_diarios['Incidence_Rate'] = np.where(base_casos_diarios['Incidence_Rate'].isna(), base_casos_diarios['Incident_Rate'], base_casos_diarios['Incidence_Rate'])
```

```python
base_casos_diarios_2= base_casos_diarios[['fecha','Country_Region','Active', 'Admin2','Case_Fatality_Ratio', 'Combined_Key', 'Confirmed','Deaths', 'FIPS',
       'Incidence_Rate', 'Last_Update', 'Latitude', 'Longitude','Province_State', 'Recovered']].copy()
```

```python
base_casos_diarios_2['fecha'] = pd.to_datetime(base_casos_diarios_2['fecha'], format ='%m-%d-%Y')
```

### Cruzando las bases

```python
base_casos_diarios_3 = base_casos_diarios_2.merge(paises ,how='left', left_on= ['Country_Region'], right_on = 'País' )
print(base_casos_diarios_2.shape)
print(base_casos_diarios_3.shape)
```
